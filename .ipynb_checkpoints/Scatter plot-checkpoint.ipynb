```python
import numpy as np
import pandas as pd
import random
import numpy as np
from Bio import SeqIO
import matplotlib.pyplot as plt
from operator import itemgetter 
from matplotlib import pyplot
from collections import Counter
import tqdm

from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

SMALL_SIZE = 8
MEDIUM_SIZE = 20
BIGGER_SIZE = 20

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
sc_path= 'ido1_feature/'
df_sfold = pd.read_csv(sc_path+'sfold.csv')
target_DB = pd.read_csv(sc_path+'ido1_exp.csv')
```

## Entire

```python
elements=['x1','x2','x3']

X = target_DB[elements]
Y = target_DB['RATE'].values
mask = ~target_DB['RATE'].isna()

X = X[mask]

Y = Y[mask]
Y = 1-Y

X_train=X
y_train=Y
X_test=X
y_test=Y
```

```python

reg = LinearRegression(fit_intercept=True).fit(X_train, y_train)
y_pred = reg.predict(X_test)
y_train_pred = reg.predict(X_train)
coef = reg.coef_/reg.coef_[0]

r_squared = r2_score(y_test, y_pred)

fig = plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred,marker = 'D', c='b',s=20)
plt.plot(np.unique(y_test), np.poly1d(np.polyfit(y_test, y_pred, 1))(np.unique(y_test)),color='k')


plt.xlabel('Experimental inhibition rate (%)')
plt.ylabel('Predicted inhibition rate (%)')
plt.xlim([0.1, 1.01])
plt.ylim([0.3,1.01])


print('Correlation: {:.4f}'.format(np.sqrt(r_squared)))
print('R2: {:.4f}'.format(r_squared))

plt.show()
```

```python
X_train_copy = X_train.copy()
X_train_copy['y'] = y_train

# dat = sm.datasets.get_rdataset("Guerry", "HistData").data
import statsmodels.formula.api as smf
# # Fit regression model (using the natural log of one of the regressors)
results = smf.ols('y ~ x1 + x2 + x3', data=X_train_copy).fit()

# # Inspect the results
print(results.summary())
print(results.pvalues)
```

## sfold-comparison

```python
sc_path= 'ido1_feature/'
df_sfold = pd.read_csv(sc_path+'sfold.csv')
target_DB = pd.read_csv(sc_path+'ido1_exp.csv')

sfolds=[]
for ii in range(len(target_DB)):
    if target_DB.iloc[ii]['len_seq']==14:
        refine = 4
    elif target_DB.iloc[ii]['len_seq']==15:
        refine = 3
    elif target_DB.iloc[ii]['len_seq']==16:
        refine = 2
    elif target_DB.iloc[ii]['len_seq']==17:
        refine = 2
                     
    sfolds.append(df_sfold.iloc[target_DB.iloc[ii]['binding_site']-4]['gibbs'])
    
target_DB['sfold'] = sfolds
target_DB = target_DB[target_DB['binding_site']!=-1]
```

```python
unique_seq=target_DB['ATGC'].unique()
random.seed('2023-08-18')
random.shuffle(unique_seq)
```

```python
train_seq = unique_seq[:int(0.80*len(unique_seq))-1]
test_seq = unique_seq[int(0.80*len(unique_seq))-1:]


target_DB_train=target_DB[target_DB['ATGC'].isin(train_seq)]
target_DB_test=target_DB[target_DB['ATGC'].isin(test_seq)]
# target_DB_test[target_DB_test['sfold']!=-1]
elements=['x1','x2','x3']

elements2=['sfold']
```

```python
len(target_DB_train), len(target_DB_test)
```

```python
X_train = target_DB_train[elements]
X_train_sfold = target_DB_train[elements2]
y_train = target_DB_train['RATE'].values

mask = ~target_DB_train['RATE'].isna()

X_train = X_train[mask]
X_train_sfold = X_train_sfold[mask]
y_train = y_train[mask]
y_train = 1-y_train

X_test = target_DB_test[elements]
X_test_sfold = target_DB_test[elements2]
y_test = target_DB_test['RATE'].values

mask = ~target_DB_test['RATE'].isna()

X_test = X_test[mask]
X_test_sfold = X_test_sfold[mask]
y_test = y_test[mask]
y_test = 1-y_test

```

```python
reg = LinearRegression(fit_intercept=True).fit(X_train, y_train)
y_pred = reg.predict(X_test)
y_train_pred = reg.predict(X_train)
coef = reg.coef_/reg.coef_[0]

r_squared = r2_score(y_test, y_pred)


reg2= LinearRegression(fit_intercept=True).fit(X_train_sfold, y_train)
y_pred_sfold = reg2.predict(X_test_sfold)
y_train_pred_sfold = reg2.predict(X_train_sfold)
coef2 = reg2.coef_/reg2.coef_[0]

r_squared_sfold = r2_score(y_test, y_pred_sfold)


fig = plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred,marker = 'D', c='b',s=20)
plt.scatter(y_test, y_pred_sfold, marker = 's', c='r',s=20)
# plt.scatter(y_train, y_train_pred,marker = 'D', c='b',s=20)
plt.plot(np.unique(y_test), np.poly1d(np.polyfit(y_test, y_pred, 1))(np.unique(y_test)),color='b')
plt.plot(np.unique(y_test), np.poly1d(np.polyfit(y_test, y_pred_sfold, 1))(np.unique(y_test)),color='r')
# plt.legend(['ASOptimizer', 'sfold'])

plt.xlabel('Experimental inhibition rate (%)')
plt.ylabel('Predicted inhibition rate (%)')
plt.xlim([0.1, 1.01])
plt.ylim([0.48,1.01])


print('Correlation: {:.4f}'.format(np.sqrt(r_squared)))
print('R2: {:.4f}'.format(r_squared))

print('Sfold: Correlation: {:.4f}'.format(np.sqrt(r_squared_sfold)))
print('Sfold: R2: {:.4f}'.format(r_squared_sfold))

plt.show()
```
